```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the database
data = pd.read_csv("C:/Users/shrey/OneDrive/Desktop/Tips.csv")

#scatter plot with day against tip
plt.plot(data['tip'])
plt.plot(data['size'])

#adding title to the plot
plt.title("scatter plot")

#setting the x and y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#reading the database
data = pd.read_csv("C:/Users/shrey/OneDrive/Desktop/Tips.csv")

#Bar Chart with day against tip
plt.bar(data['day'], data['tip'])

plt.title("Bar Chart")

#setting the x and y labels
plt.xlabel('Day')
plt.ylabel('Tip')

#adding the legends
plt.show()
```
